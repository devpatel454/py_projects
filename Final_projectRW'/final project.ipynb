```python
print("===== final project =====")

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("2015.csv")

# Display basic info
print(df.info())  # Check column names, data types, and missing values
print(df.head())  # Preview the first five rows



# Check for missing values
print(df.isnull().sum())

# Rename columns
df.rename(columns={"Happiness Score": "Score"}, inplace=True)



# Histogram of a numerical column
plt.hist(df["Score"], bins=20, color="skyblue", edgecolor="black")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.title("Distribution of happiness Scores")
plt.show()





# bar plot top 10 countries by happiness score
plt.figure(figsize=(12, 6))
top_10 = df.nlargest(10, 'Score')
sns.barplot(x='Score', y='Country', data=top_10, palette="Blues_r")

plt.xlabel("Happiness Score")
plt.ylabel("Country")
plt.title("Top 10 Happiest Countries (2015)")
plt.show()


#scatter plot (happiness score vs gdp per capita)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Economy (GDP per Capita)', y='Score', data=df, hue='Region', palette="coolwarm")

plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Score")
plt.title("Happiness Score vs GDP per Capita")
plt.legend(title="Region", bbox_to_anchor=(1, 1))
plt.show()



# box plot (happiness score by region)
plt.figure(figsize=(12, 6))
sns.boxplot(x='Region', y='Score', data=df)
plt.xticks(rotation=90)

plt.xlabel("Region")
plt.ylabel("Happiness Score")
plt.title("Happiness Score by Region")
plt.show()



# scatter plot : comparing gdp and life expectancy for happiness

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Economy (GDP per Capita)'], y=df['Health (Life Expectancy)'], hue=df['Score'], size=df['Score'], palette='coolwarm', sizes=(20, 200))

plt.xlabel('GDP per Capita')
plt.ylabel('Health (Life Expectancy)')
plt.title('Comparison: GDP vs. Life Expectancy with Happiness Score')
plt.show()


# Average Happiness Score by Region
region_avg = df.groupby("Region")["Score"].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=region_avg.values, y=region_avg.index, palette="Blues_r")
plt.xlabel("Average Happiness Score")
plt.ylabel("Region")
plt.title("Average Happiness Score by Region")
plt.show()





#  Violin Plot: Regional Happiness Distribution
plt.figure(figsize=(12, 6))
sns.violinplot(x='Region', y='Score', data=df, palette="coolwarm")
plt.xticks(rotation=90)
plt.xlabel("Region")
plt.ylabel("Happiness Score")
plt.title("Happiness Score Distribution by Region")
plt.show()


# top 10 least happiness countries

plt.figure(figsize=(12, 6))
bottom_10 = df.nsmallest(10, 'Score')
sns.barplot(x='Score', y='Country', data=bottom_10, palette="Reds_r")
plt.xlabel("Happiness Score")
plt.ylabel("Country")
plt.title("Bottom 10 Least Happy Countries (2015)")
plt.show()
























```
